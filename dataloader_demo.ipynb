# Demo for dataloader.py

This notebook demonstrates how to use the methods in dataloader.py

```python
# First import the module
import dataloader
```

```python
# Then init the module
dataloader.init()
```

```python
# You can get a list of all the unique series_ids in the data
sids = dataloader.all_series_ids()
sids
# The result also gets cached in the function so that it will only be slow the first time you call it
```

```python
# You can then get the data for a particular series_id like so
acc3 = dataloader.acc_data_for_child(sids[3])
# This took me 3m 33s to load the first time
```

```python
# You can use the dataframe normally
acc3
```

```python
# The next time you load the same data it will be faster
acc3_ = dataloader.acc_data_for_child(sids[3])
# This time it took only 0.5
```
